```python
import pandas as pd
df_jugadores = pd.read_csv('../jugadores_cebollitas.csv')
df_jugadores.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df_jugadores[['goles', 'asistencias', 'pases_completados (%)', 'tiros_al_arco']])
plt.show()
```

```python
from sklearn.cluster import KMeans

X = df_jugadores[['goles', 'asistencias', 'pases_completados (%)', 'tiros_al_arco']]

kmeans = KMeans(n_clusters=3, random_state=42)
df_jugadores['cluster'] = kmeans.fit_predict(X)

df_jugadores.head()
```

```python
sns.scatterplot(data=df_jugadores, x='goles', y='asistencias', hue='cluster')
plt.title("Clusters de jugadores por goles y asistencias")
plt.xlabel("Goles anotados")
plt.ylabel("Asistencias")
plt.show()
```

```python
columnas_metricas = ['goles', 'asistencias', 'pases_completados (%)', 'tiros_al_arco']
perfiles = df_jugadores.groupby('cluster')[columnas_metricas].mean()
print(perfiles)
```

```python
import ipywidgets as widgets

def clustering_interactive(k):
    km = KMeans(n_clusters=k, random_state=42)
    df_jugadores['cluster'] = km.fit_predict(X)

    columnas_metricas = ['goles', 'asistencias', 'pases_completados (%)', 'tiros_al_arco']
    perfiles = df_jugadores.groupby('cluster')[columnas_metricas].mean()
    print(perfiles)
    
    sns.scatterplot(data=df_jugadores, x='goles', y='asistencias', hue='cluster', palette='Set1')
    plt.title(f"Clusters de jugadores con k={k}")
    plt.xlabel("Goles anotados")
    plt.ylabel("Asistencias")
    plt.show()

widgets.interactive(clustering_interactive, k=widgets.IntSlider(min=2, max=6, step=1, value=3))
```
